```python
Xin = [0.3432,0,0.6667,0.5678,0.473,0,0.5,0.6794,1,0.3456,1,0.25,1]
```

```python
X.shape[0]
```

```python
def dist(x,y):
  distance = np.linalg.norm(x - y)
  return distance
```

```python
print(Xin,X.iloc[1])
```

```python
distance  = []
for i in range(X.shape[0]):
  distance.append(dist(Xin,X.iloc[i]))
```

```python
K = 3
```

```python
dis = np.array(distance)
```

```python
array1 = np.sort(dis)
```

```python
array1
```

```python
array1[0]
```

```python
least_dis=[]
for i in range(K):
  least_dis.append(array1[i])
```

```python
least_dis
```

```python
least_index=[]
```

```python
for j in range(K):
   for i in range(303):
       if (distance[i]==least_dis[j]):
           least_index.append(i)

```

```python
least_index
```

```python
prediction =0
for i in least_index:
  prediction = prediction + Y[i]
prediction = prediction/K
```

```python
prediction
```

```python
if prediction>0.5:
  target = 1
else:
  target= 0
```

```python
from scipy.spatial import Voronoi, voronoi_plot_2d
vor = Voronoi(X)
```

```python
voronoi_plot_2d(vor, show_vertices=False, line_colors='orange', line_width=2)
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired, edgecolor='K', s=20)
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title('Voronoi Diagram with KNN Decision Boundaries')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
